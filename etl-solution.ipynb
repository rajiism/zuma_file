Data Transformation

```python
#replacing missing value

real_estate_df.fillna({
    'assessorID':'Uknown',
    'bedrooms':0.0,
    'features':'unknown',
    'legalDescription':'unknown',
    'ownerOccupied':0.0,
    'squareFootage':0.0,
    'yearBuilt':0.0,
    'zoning':'unknown',
    'bathrooms':0.0,
    'lotSize':0.0,
    'propertyType':'unknown',
    'taxAssessment':'unknown',
    'propertyTaxes':'unknown',
    'lastSalePrice':0.0,
    'owner':'unknown',
    'addressLine2':'unknown',
    'subdivision':'unknown'
}, inplace=True)
```

```python
real_estate_df.info()
```

Creating the tables

```python
location_dim = real_estate_df[['addressLine1', 'city', 'state', 'zipCode', 'formattedAddress',
                               'county','longitude', 'latitude','addressLine2']].copy().drop_duplicates().reset_index(drop=True)

location_dim.index.name = 'location_id'
```

```python
location_dim = location_dim.reset_index()


```

```python
location_dim.head()
```

```python
sales_dim = real_estate_df[['lastSalePrice', 'lastSaleDate']].copy().drop_duplicates().reset_index(drop=True)

sales_dim.index.name = 'sales_id'

sales_dim
```

```python
sales_dim=sales_dim.reset_index()

sales_dim.head()
```

```python
#converting features to string
real_estate_df['features'] = real_estate_df['features'].astype(str) 
```

```python
real_estate_df['propertyTaxes'] = real_estate_df['propertyTaxes'].astype(str) 
real_estate_df['taxAssessment'] = real_estate_df['taxAssessment'].astype(str) 
```

```python
features_dim = real_estate_df[['bedrooms', 'bathrooms', 'lotSize','features','squareFootage']].copy().drop_duplicates().reset_index(drop=True)

features_dim.index.name = 'features_id'
```

```python
features_dim=features_dim.reset_index()
```

```python
features_dim.head()
```

```python
fact_real_estate = real_estate_df.merge(sales_dim, on=['lastSalePrice', 'lastSaleDate'], how='left')\
                                 .merge(location_dim, on=['addressLine1', 'city', 'state', 'zipCode', 'formattedAddress','county','longitude', 'latitude','addressLine2'], how='left')\
                                 .merge(features_dim, on = ['bedrooms', 'bathrooms', 'lotSize','features','squareFootage'], how = 'left')\
                                 [['id','location_id','sales_id','features_id','assessorID','legalDescription','ownerOccupied','yearBuilt', 'zoning', 'propertyType', 'taxAssessment', 'propertyTaxes','subdivision']]
```

```python
def create_table():
    conn = get_db_connection()
    cursor=conn.cursor()
    create_table_query = '''
                            create schema if not exists realtors;

                            drop table if exists realtors.location_dim cascade;
                            drop table if exists realtors.sales_dim cascade;
                            drop table if exists realtors.features_dim cascade;
                            drop table if exists realtors.fact_real_estate cascade;

                            create table realtors.location_dim(
                                location_id int primary key,
                                addressLine1 varchar(100000),
                                city varchar(100000),
                                state varchar(100000),
                                zipCode int,
                                formattedAddress varchar(100000),
                                county varchar(100000),
                                longitude float,
                                latitude float,
                                addressLine2 varchar(100000)
                            );

                                create table realtors.sales_dim(
                                sales_id int primary key,
                                lastSalePrice float,
                                lastSaleDate varchar(100000)
                                
                            );
                                create table realtors.features_dim(
                                features_id int primary key, 
                                bedrooms float, 
                                bathrooms float, 
                                lotSize float,
                                features varchar(100000),
                                squareFootage float
                                                               
                            );

                                create table realtors.fact_real_estate(
                                id varchar(100000)primary key,
                                location_id int,
                                sales_id int,
                                features_id int,
                                assessorID varchar(100000),
                                legalDescription varchar(100000),
                                ownerOccupied float, 
                                yearBuilt float,
                                zoning varchar(100000),
                                propertyType varchar(100000),
                                taxAssessment varchar(100000),
                                propertyTaxes varchar(100000),
                                subdivision varchar(100000),
                                foreign key(location_id) references realtors.location_dim(location_id),
                                foreign key(sales_id) references realtors.sales_dim(sales_id),
                                foreign key(features_id) references realtors.features_dim(features_id)
                            );
                                                      
                            '''
    cursor.execute(create_table_query)
    conn.commit()
    cursor.close()
    conn.close()
```
